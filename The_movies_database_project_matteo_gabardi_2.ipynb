#### 4.Extract the domain of each homepage.

```python
# To begin with, it is important to notice that one movie has more than one homepage.
movies[movies['homepage'].str.contains(' ', na=False)]
```

```python
# Function that extracts domains from 'homepage'.
def domain_extractor (row):
    aux_string=""
    row=row.split(" ")
    
    # If more than one homepage is present, the loop allows to treat them singularly.
    for aux in range(len(row)):
        
        # Protocol is removed.
        row[aux]=row[aux].split("//")[1].split("/")[0]

        # 'www.' is removed if present
        if(row[aux].startswith('www.')==True):
            row[aux]=row[aux][4:]
            
        # If it is the first homepage
        if aux==0:        
             aux_string=row[aux]
                
        # If it is not the first homepage, the following domains are added with " " as separator.
        else:
            aux_string=aux_string+" "+row[aux]

    return aux_string


# Function that extracts top-level domains from 'homepage'.
def top_level_domain_extractor (row):
    aux_string=""
    row=row.split(" ")
    
    # If more than one domain is present, the loop allows to treat them singularly.
    for aux in range(len(row)):
        
        # If it is the first domain.
        if aux==0:        
            # The top-level domain is separated.
            aux_string=row[aux].split(".")[-1]
            
        # If it is not the first domain, the following top-level domains are added with " " as separator.
        else:
            aux_string=aux_string+" "+row[aux].split(".")[-1]

    return aux_string


# The domain is extracted applying 'domain_extractor' to all rows.
movies['domain']=movies[movies['homepage'].notnull()]['homepage'].apply(domain_extractor)

# Top-level domain is extracted.
movies['top-level domain']=movies[movies['homepage'].notnull()]['domain'].apply(top_level_domain_extractor)

movies[movies['homepage'].notnull()][['title','homepage','domain','top-level domain']]
```

#### 5. Extract a set of normalized tables.

###### Definition of the functions used

```python

# Function that reads the json attribute for each row with pd.read_json, adds the other attributes of the row and append the resultant row to the input dataframe.
def json_to_df(row, dfs, json_col):
    json_df = pd.read_json(row[json_col])
    dfs.append(json_df.assign(**row.drop(json_col)))

# Function that from an empty list returns the normalized table using the json_to_df function that fill the list.
def table_normalizer(table, json_col):
    aux_ls = [] # Empty auxiliary list
    table.apply(json_to_df, json_col=json_col, dfs=aux_ls, axis=1) # Apply the function json_to_df to all rows
    return pd.concat(aux_ls) # The result is a dataframe
```

###### Normalization

```python
#Normalization of 'crew' json attribute in credits.
crew_norm=credits[['crew','title','movie_id']]# Attributes to be inserted in the table are selected.
crew_norm=table_normalizer(crew_norm, json_col='crew')# Normalization based on the defined json column.
crew_norm.replace({"gender": {2.0: "M", 1.0: "F", 0.0: None}}, inplace = True)# 'gender' attribute is redefined: 2.0 corresponds to M (male), 1.0 corresponds to F (female), 0.0 corresponds to  None (not defined).
crew_norm.set_index('credit_id', inplace=True, verify_integrity=True) # Setting the index, controlling the uniqueness of the indexing.
crew_norm.rename(columns={'id': 'crew id'}, inplace=True) 
crew_norm.head()
```

```python
#Normalization of 'cast' json attribute in credits.
cast_norm=credits[['cast','title','movie_id']]
cast_norm=table_normalizer(cast_norm, json_col='cast')
cast_norm.replace({"gender": {2.0: "M", 1.0: "F", 0.0: None}}, inplace = True)
cast_norm.set_index('credit_id', inplace=True, verify_integrity=True) 
cast_norm.rename(columns={'id': 'actor id', 'name': 'actor name'}, inplace=True)
cast_norm.head()
```

```python
#Normalization of 'genres' json attribute in movies.
genres_norm=movies[['genres','title','movie_id']]
genres_norm=table_normalizer(genres_norm, json_col='genres')
genres_norm.rename(columns={'id': 'genre id', 'name':'genre'}, inplace=True)
genres_norm.set_index(['genre id','movie_id'], inplace=True, verify_integrity=True)
genres_norm.head()
```

```python
#Normalization of 'keywords' json attribute in movies.
keywords_norm=movies[['keywords','title','movie_id']]
keywords_norm=table_normalizer(keywords_norm, json_col='keywords')
keywords_norm.rename(columns={'id': 'keyword id', 'name': 'keyword name'}, inplace=True)
keywords_norm.set_index(['keyword id','movie_id'], inplace=True, verify_integrity=True)
keywords_norm.head()
```

```python
#Normalization of 'production_companies' json attribute in movies.
prod_comp_norm=movies[['production_companies','title','movie_id']]
prod_comp_norm=table_normalizer(prod_comp_norm, json_col='production_companies')
prod_comp_norm.rename(columns={'name': 'production company','id': 'company id'}, inplace=True)
prod_comp_norm.set_index(['company id','movie_id'], inplace=True, verify_integrity=True)
prod_comp_norm.head()
```

```python
#Normalization of 'production_countries' json attribute in movies.
prod_country_norm=movies[['production_countries','title','movie_id']]
prod_country_norm=table_normalizer(prod_country_norm, json_col='production_countries')
prod_country_norm.rename(columns={'name': 'country'}, inplace=True)
prod_country_norm.set_index(['movie_id','iso_3166_1'], inplace=True, verify_integrity=True)
prod_country_norm.head()
```

```python
#Normalization of 'spoken_languages' json attribute in movies.
spoken_language_norm=movies[['spoken_languages','title','movie_id']]
spoken_language_norm=table_normalizer(spoken_language_norm, json_col='spoken_languages')
spoken_language_norm.rename(columns={'name': 'language'}, inplace=True)
spoken_language_norm.set_index(['movie_id','iso_639_1'], inplace=True, verify_integrity=True)
spoken_language_norm.head()
```

#### 10.For each company and each decade, compute the overall revenue
